### Introduction
In this Assignment, we aim to demonstrate the use of autoencoders for image reconstruction, specifically focusing on custom character images ( R ) in addition to the MNIST dataset.

### Description
The project involves the following steps:
1. Loading the Pre-trained Autoencoder and Preprocessing the custom character images: We load and preprocess grayscale images of custom characters, ensuring they are of the same size as the MNIST dataset (28x28 pixels).
2. Oversampling the custom character images: To incorporate the custom characters into the training data, we oversample each custom character image multiple times to match the size of the MNIST dataset. This ensures that the autoencoder model can effectively learn representations of the custom characters.
3. Combining the oversampled custom character images with the MNIST dataset: We concatenate the oversampled custom character images with the MNIST training dataset to create a combined dataset for training the autoencoder.
4. Training the autoencoder: We train the autoencoder model using the combined dataset, where the model learns to reconstruct both MNIST digits and the custom characters.
5. Evaluating the autoencoder: We evaluate the performance of the trained autoencoder by reconstructing random samples from the combined dataset and visually comparing the original and reconstructed images.
6. Testing the autoencoder on new custom character images: We preprocess and predict the reconstructed images for a separate set of custom character images to demonstrate the model's ability to reconstruct unseen custom characters.

### Loading Libraries and Modules

```python
import numpy as np
from matplotlib import pyplot as plt
from tensorflow import keras
import tensorflow as tf
 
from tensorflow.keras.callbacks import EarlyStopping 
```

### Loading Pre-trained Digit Autoencoder

```python
pretrained_model = keras.models.load_model('initially_trained_autoencoder.h5')
```

### Load and Process the MNIST Dataset

```python
from tensorflow.keras.datasets import mnist
# load MNIST data
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print(train_images.shape)
print(test_images.shape)

# Flatten the images for a autoencoder input (28x28 is 784 -- therfore, input is 784 neurons)
train_images = np.reshape(train_images, (-1, 784))
test_images = np.reshape(test_images, (-1, 784))
print(train_images.shape)
print(test_images.shape)

# Normalize data - since it's grayscale, we can divide by 255 to get each pixel between 0 and 1
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255
```

### Check the Pre-trained model results

```python
reconstructed_images = pretrained_model.predict(test_images)
images = []
for i in range(100):
    images.append(test_images[i].reshape(28,28))
    images.append(reconstructed_images[i].reshape(28,28))
rows = 10
cols = 10
plt.figure(figsize=(10,10))
for i in range(rows*cols):
    plt.subplot(rows,cols,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    if (i%2 == 0):
        plt.imshow(images[i],cmap='gray')
    else:
        plt.imshow(images[i])
plt.show()
```

### Testing the model on our 5 Images

```python
from matplotlib import pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
 
 
 
def preprocess_image(image_path):
    img = tf.keras.utils.load_img(image_path, target_size=(28, 28), color_mode="grayscale")
    img_array = tf.keras.utils.img_to_array(img)
    img_array = img_array / 255.0  # Normalize between 0 and 1
    img_array = img_array.reshape(784)
    img_array = np.array([img_array])  # Convert single image to a batch with one image in it
    return img_array
 
# Load and preprocess 5 images
image_paths = ['testimg01.png', 'testimg02.png', 'testimg03.png', 'testimg04.png', 'testimg05.png']
images = [preprocess_image(image_path) for image_path in image_paths]
 
# Predict images using the pre-trained autoencoder model
predictions = [pretrained_model.predict(image) for image in images]
 
 
# Display the images side by side
plt.figure(figsize=(15, 5))
for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(images[i].reshape(28, 28), cmap="gray")
    plt.subplot(2, 5, i+6)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(predictions[i].reshape(28, 28), cmap="gray")
 
plt.show()
```

We Can see in the above results that our pretrained moel performs poorly on the character ( R )

### Oversample the Character ( R ) Images

```python
# Reshape each image to a 784-dimensional vector
images_reshaped = [img.reshape(784) for img in images]
# Replicate each image multiple times to create 1000 samples in total
num_samples_per_image = 1000
new_images = np.concatenate([np.repeat(img[np.newaxis, :], num_samples_per_image, axis=0) for img in images_reshaped], axis=0)
# Ensure the total number of images is exactly 5000
new_images = new_images[:5000]
# Shape of the new oversampled images
print(new_images.shape)  # Should print (5000, 784)
```

```python
# Concatenate oversampled images with MNIST dataset
train_images_combined = np.concatenate([train_images, new_images], axis=0)

# Print shapes of the combined datasets
print("Combined Train Images Shape:", train_images_combined.shape)
print("Test Images Shape:", test_images.shape)
```

### Creating a new Autoencoder and fit on our new data

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
input_dim = 28*28
latent_vec_len = 20 # this is the 'bottleneck' of the autoencoder
# Define the autoencoder architecture
# build the encoder
autoencoder = Sequential()
autoencoder.add(Input(shape=(input_dim,)))
autoencoder.add(Dense(600, activation='relu'))
autoencoder.add(Dense(400, activation='relu'))
autoencoder.add(Dense(100, activation='relu'))

# the bottleneck layer
autoencoder.add(Dense(latent_vec_len, activation='relu'))

# the decoder layer
autoencoder.add(Dense(100, activation='relu'))
autoencoder.add(Dense(400, activation='relu'))
autoencoder.add(Dense(600, activation='relu'))
autoencoder.add(Dense(input_dim, activation='sigmoid'))

# Compile the autoencoder model
autoencoder.compile(loss='binary_crossentropy', optimizer='adam')

# Get summary
autoencoder.summary()
```

```python
# EarlyStopping during model training
# Train the autoencoder with MNIST data
early_stop = EarlyStopping(monitor='loss', patience=3, verbose=1)
history = autoencoder.fit(
    x=train_images_combined, # For autoencoders, both the input and the target are the same
    y=train_images_combined, 
    epochs=12, 
    batch_size=128,
    shuffle=True, 
    validation_data=(test_images, test_images),
    callbacks=[early_stop]
)
```

```python
# Plot training and validation loss scores
# against the number of epochs.
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.ylabel('Binary Cross Entropy Loss')
plt.xlabel('Epoch')
plt.title('Autoencoder Reconstruction Loss', pad=13)
plt.legend(loc='upper right')
```

### Predicting and displaying each character image 

```python
# Predict and display each image
num_images = len(image_paths)
plt.figure(figsize=(10, 4))

for i, image_path in enumerate(image_paths, start=1):
    input_arr = preprocess_image(image_path)
    predictions = autoencoder.predict(input_arr)

    plt.subplot(2, num_images, i)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(input_arr.reshape(28, 28), cmap="gray")
    plt.xlabel('Original')

    plt.subplot(2, num_images, i + num_images)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(predictions[0].reshape(28, 28), cmap="gray")
    plt.xlabel('Predicted')

plt.tight_layout()
plt.show()
```
